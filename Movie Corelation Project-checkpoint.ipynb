```python
# importing libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now read in the data
df = pd.read_csv(r'C:\Users\tp.admin\Downloads\movies.csv')
```

```python
# display the data

df.head()
```

```python
# check if there is any missing data 
# Looping through the data to look for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Look for outliers in the data

df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
# Order Data a little bit to see

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# to view all the data in the tables, do not need to write the code again to display. Run the above line again and will display all the data

pd.set_option('display.max_rows', None)
```

#### finding the corelation between the data

```python
# scatter plot budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn to confirm the corelation

sns.regplot(x="gross", y="budget", data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# this is the default corelation, all three gives different values

df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes
        
df_numeric.head()
```

```python
correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numeric.corr()
```

```python
corelation_mat = df_numeric.corr()

corr_pairs = corelation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
